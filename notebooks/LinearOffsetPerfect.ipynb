```python
import matplotlib.pyplot as plt
import numpy as np
import ast, random, dataIO
from sklearn import linear_model, preprocessing, cross_validation
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
reload(dataIO)

def get_touch_locations(userid):
    left = {}
    right = {}
    filenos = ["2", "6", "9"]
    
    for fileno in filenos:
        filename = "/home/dimitar/Desktop/Python/experiment/results/"+str(userid)+"_"+fileno+"up.txt"
        with open(filename, "r") as f:
            lines = f.read().splitlines()
            lines = map(lambda x: x.split('\t'), lines[1:])
            map(lambda x: left.setdefault(x[1],[]).append(ast.literal_eval(x[3])) if x[0]=="left" 
                else right.setdefault(x[1],[]).append(ast.literal_eval(x[3])), lines)
        
    return left, right
    
left, right = get_touch_locations(17)
centers = dataIO.get_key_centers()
left_filtered = dataIO.filter_touches(left)
right_filtered = dataIO.filter_touches(right)

points = []
left_train = []
left_test = []
t_l_x_train = []
t_l_x_test = []
t_l_y_train = []
t_l_y_test = []

right_train = []
right_test = []
t_r_x_train = []
t_r_x_test = []
t_r_y_train = []
t_r_y_test = []

keys = left_filtered.keys()
touch_centers = []
        
for key in keys:
    center = centers[key]
    left_touches = left_filtered[key]
    right_touches = right_filtered[key]
    #random.shuffle(left_touches)
    #random.shuffle(right_touches)
    
    for i in range(len(left_touches)):
        touch = left_touches[i]
        touch_centers.append(center)
        points.append(list(touch))      
        
        if i>3:
            left_test.append(list(touch))
            t_l_x_test.append(center[0]-touch[0])
            t_l_y_test.append(center[1]-touch[1])            
        else:
            left_train.append(list(touch))
            t_l_x_train.append(center[0]-touch[0])
            t_l_y_train.append(center[1]-touch[1])
            
    for i in range(len(right_touches)):
        touch = right_touches[i]
        touch_centers.append(center)
        points.append(list(touch))      
        
        if i>3:
            right_test.append(list(touch))
            t_r_x_test.append(center[0]-touch[0])
            t_r_y_test.append(center[1]-touch[1])            
        else:
            right_train.append(list(touch))
            t_r_x_train.append(center[0]-touch[0])
            t_r_y_train.append(center[1]-touch[1])
 
touch_centers = np.array(touch_centers)
points = np.array(points)

t_l_x_train = np.array(t_l_x_train)
t_l_x_test = np.array(t_l_x_test)
t_l_y_train = np.array(t_l_y_train)
t_l_y_test = np.array(t_l_y_test)

t_r_x_train = np.array(t_r_x_train)
t_r_x_test = np.array(t_r_x_test)
t_r_y_train = np.array(t_r_y_train)
t_r_y_test = np.array(t_r_y_test)

points_unz = map(lambda x: np.array(x) , zip(*points))
centers_unz = map(lambda x: np.array(x) , zip(*touch_centers))

print("MSE on x before regression: %.2f"
           % np.mean((points_unz[0] - centers_unz[0]) ** 2))

print("MSE squared error on y before regression: %.2f"
           % np.mean((points_unz[1] - centers_unz[1]) ** 2))
print

regr_l_x = linear_model.LinearRegression()
regr_l_y = linear_model.LinearRegression()

regr_l_x.fit(left_train, t_l_x_train)
regr_l_y.fit(left_train, t_l_y_train)

true, pred = t_l_x_test, regr_l_x.predict(left_test)

mse_x=[]
mse_y=[]

mse = np.mean((pred - true) ** 2)
print("MSE on left thumb on x: %.2f" %mse)
mse_x.append(mse)

true, pred = t_l_y_test, regr_l_y.predict(left_test)

mse = np.mean((pred - true) ** 2)
print("MSE on left thumb on y: %.2f" %mse)
mse_y.append(mse)

regr_r_x = linear_model.LinearRegression()
regr_r_y = linear_model.LinearRegression()

regr_r_x.fit(right_train, t_r_x_train)
regr_r_y.fit(right_train, t_r_y_train)

true, pred = t_r_x_test, regr_r_x.predict(right_test)

mse = np.mean((pred - true) ** 2)
print("MSE on right thumb on x: %.2f" %mse)
mse_x.append(mse)

true, pred = t_r_y_test, regr_r_y.predict(right_test)

mse = np.mean((pred - true) ** 2)
print("MSE on right thumb on y: %.2f" %mse)
mse_y.append(mse)

print
print("Mean MSE on on x: %.2f" %np.mean(np.array(mse_x)))
print("Mean MSE on on y: %.2f" %np.mean(np.array(mse_y))) 

left_unz = zip(*left_test)
right_unz = zip(*right_test)
```
