```python
import matplotlib.pyplot as plt
import numpy as np
import ast, random, dataIO
from sklearn import linear_model, cross_validation
%matplotlib inline

def get_touch_locations(userid):
    data = {}
    filenos = ["2", "6", "9"]
    
    for fileno in filenos:
        filename = "/home/dimitar/Desktop/Python/experiment/results/"+str(userid)+"_"+fileno+"up.txt"
        with open(filename, "r") as f:
            lines = f.read().splitlines()
            lines = map(lambda x: x.split('\t'), lines[1:])
            map(lambda x: data.setdefault(x[1],[]).append(ast.literal_eval(x[3])), lines)
        
    return data
    
locations = get_touch_locations(4)
centers = dataIO.get_key_centers()
filtered = dataIO.filter_touches(locations)

points = []
targets_x = []
targets_y = []
keys = filtered.keys()
touch_centers = []
        
for key in keys:
    touches = filtered[key]
    center = centers[key]
    
    for i in range(len(touches)):
        touch = touches[i]
        touch_centers.append(center)
        points.append(list(touch))
        #points.append(list(touch)+[touch[0]**2, touch[1]**2])
        targets_x.append(center[0]-touch[0])
        targets_y.append(center[1]-touch[1])       
        
touch_centers = np.array(touch_centers)
points = np.array(points)

targets_x = np.array(targets_x)
targets_y = np.array(targets_y)

points_unz = np.array(map(lambda x: list(x) , zip(*points))) 
centers_unz = np.array(map(lambda x: list(x) , zip(*touch_centers)))

print("Mean squared error on x before regression: %.2f"
           % np.mean((points_unz[0] - centers_unz[0]) ** 2))

print("Mean squared error on y before regression: %.2f"
           % np.mean((points_unz[1] - centers_unz[1]) ** 2))

foldno = 1
mse_x = []
mse_y = []
kf = cross_validation.KFold(len(points), n_folds=10, shuffle=True)

for train_index, test_index in kf:
    print
    print ("##### Fold %d #####" %foldno)
    print
    
    points_train, points_test = points[train_index], points[test_index]
    t_x_train, t_x_test = centers_unz[0][train_index], centers_unz[0][test_index]
    t_y_train, t_y_test = centers_unz[1][train_index], centers_unz[1][test_index]
    
    regr_x = linear_model.LinearRegression()
    regr_y = linear_model.LinearRegression()
    
    regr_x.fit(points_train, t_x_train)
    regr_y.fit(points_train, t_y_train)
    
    true, pred = t_x_test, regr_x.predict(points_test)
    
    mse = np.mean((pred - true) ** 2)
    print("Mean squared error on x: %.2f" %mse)
    mse_x.append(mse)
    
    true, pred = t_y_test, regr_y.predict(points_test)
    
    mse = np.mean((pred - true) ** 2)
    print("Mean squared error on y: %.2f" %mse)
    mse_y.append(mse)
    foldno +=1

print
print "####################"
print 
print("Mean MSE on x: %.2f" %np.mean(np.array(mse_x)))
print("Mean MSE on y: %.2f" %np.mean(np.array(mse_y)))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import ast, dataIO
from sklearn import linear_model, cross_validation
%matplotlib inline
reload(dataIO)

def get_touch_locations(userid):
    touches = []
    filenos = ["2", "6", "9"]
    
    for fileno in filenos:
        filename = "/home/dimitar/Desktop/Python/experiment/results/"+str(userid)+"_"+fileno+"down.txt"
        with open(filename, "r") as f:
            lines = f.read().splitlines()
            lines = map(lambda x: x.split('\t'), lines[1:])
            for line in lines:
                letter = line[1]
                location = ast.literal_eval(line[3])
                x = location[0]
                y = location[1]
                time = line[2]
                bod = dataIO.createlist(line[-1])
                if line[0] == "left":
                    touch = dataIO.Touch(x,y,bod,letter,time, True)
                else:
                    touch = dataIO.Touch(x,y,bod,letter,time, False)
                touches.append(touch)
        
    return touches

locations = get_touch_locations(4)
filtered = dataIO.filter_new(locations)

centers = dataIO.get_key_centers()
keys = centers.keys()

touch_centers = []
touches = []
points = []
targets_x = []
targets_y = []
labels = []

for key in keys:
    if key=="SPACE":
        continue
    
    center = centers[key]
    touch_list = dataIO.find_touches(filtered, key)
    
    for touch in touch_list:
        touch_centers.append(center)
        location = [touch.x, touch.y]
        
        points.append(location)
        touches.append(touch)
        
        targets_x.append(center[0]-location[0])
        targets_y.append(center[1]-location[1])
        labels.append(touch.letter+str(touch.left))
            
touch_centers = np.array(touch_centers)
points = np.array(points)
touches = np.array(touches)
targets_x = np.array(targets_x)
targets_y = np.array(targets_y)

points_unz = np.array(map(lambda x: list(x) , zip(*points))) 
centers_unz = np.array(map(lambda x: list(x) , zip(*touch_centers)))

print("MSE on x before regression: %.2f"
           % np.mean((points_unz[0] - centers_unz[0]) ** 2))

print("MSE on y before regression: %.2f"
           % np.mean((points_unz[1] - centers_unz[1]) ** 2))
print

foldno = 1
mse_x = []
mse_y = []
kf = cross_validation.StratifiedKFold(labels, n_folds= 5, shuffle=False)
#kf = cross_validation.KFold(len(points), n_folds=5, shuffle=False)

for train_index, test_index in kf:
    print
    print ("##### Fold %d #####" %foldno)
    print
    
    left_train = []
    right_train = []    
    left_test = []
    right_test = []
    
    t_l_x_train = []
    t_l_y_train = []
    t_l_x_test = []
    t_l_y_test = []
    
    t_r_x_train = []
    t_r_y_train = []
    t_r_x_test = []
    t_r_y_test = []
    
    for index in train_index:
        touch = touches[index]
        if touch.left:
            left_train.append([touch.x, touch.y])
            #left_train.append([touch.x, touch.y, touch.x**2, touch.y**2])
            t_l_x_train.append(targets_x[index])
            t_l_y_train.append(targets_y[index])
        else:
            right_train.append([touch.x, touch.y])
            #right_train.append([touch.x, touch.y, touch.x**2, touch.y**2])
            t_r_x_train.append(targets_x[index])
            t_r_y_train.append(targets_y[index])
            
    for index in test_index:
        touch = touches[index]
        if touch.left:
            left_test.append([touch.x, touch.y])
            #left_test.append([touch.x, touch.y, touch.x**2, touch.y**2])
            t_l_x_test.append(targets_x[index])
            t_l_y_test.append(targets_y[index])
        else:
            right_test.append([touch.x, touch.y])
            #right_test.append([touch.x, touch.y, touch.x**2, touch.y**2])
            t_r_x_test.append(targets_x[index])
            t_r_y_test.append(targets_y[index])
    
    regr_l_x = linear_model.LinearRegression()
    regr_l_y = linear_model.LinearRegression()
    
    regr_l_x.fit(left_train, t_l_x_train)
    regr_l_y.fit(left_train, t_l_y_train)
    
    true, pred = t_l_x_test, regr_l_x.predict(left_test)
    
    mse = np.mean((pred - true) ** 2)
    print("MSE on left thumb on x: %.2f" %mse)
    mse_x.append(mse)
    
    true, pred = t_l_y_test, regr_l_y.predict(left_test)
    
    mse = np.mean((pred - true) ** 2)
    print("MSE on left thumb on y: %.2f" %mse)
    mse_y.append(mse)
    
    regr_r_x = linear_model.LinearRegression()
    regr_r_y = linear_model.LinearRegression()
    
    regr_r_x.fit(right_train, t_r_x_train)
    regr_r_y.fit(right_train, t_r_y_train)
    
    true, pred = t_r_x_test, regr_r_x.predict(right_test)
    
    mse = np.mean((pred - true) ** 2)
    print("MSE on right thumb on x: %.2f" %mse)
    mse_x.append(mse)
    
    true, pred = t_r_y_test, regr_r_y.predict(right_test)
    
    mse = np.mean((pred - true) ** 2)
    print("MSE on right thumb on y: %.2f" %mse)
    mse_y.append(mse)
    
    if foldno == 1:
        list1 = right_train
        list2 = t_r_x_train
        list3 = right_test
        list4 = t_r_x_test
        regr1 = regr_r_x
        regr2 = regr_r_y
        
    foldno +=1


print 
print("MSE on on x: %.2f" % np.mean(np.array(mse_x)))
print("MSE on on y: %.2f" % np.mean(np.array(mse_y)))
```
