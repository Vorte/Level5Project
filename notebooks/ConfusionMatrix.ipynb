```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, preprocessing, cross_validation
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import confusion_matrix
%matplotlib inline

postures = {"left_hand":["4", "8", "11"], "right_hand":["1", "7", "10"], 
            "index_finger":["3", "5", "12"], "two_hand":["2", "6", "9"]}

def readfile(userid, posture):
    # TODO: up/down switching
    data = []
    filenos = postures[posture]
    
    for fileno in filenos:
        filename = "/home/dimitar/Desktop/Python/experiment/results/"+str(userid)+"_"+fileno+"down.txt"
        with open(filename, "r") as f:
            lines = f.read().splitlines()
            lines = map(lambda x: (x.split('\t')[-1]).replace('(', '').replace(')', ''), lines[1:])
            map(lambda x: data.append(map(int,x.split(', '))), lines)
        
    return np.array(data)

twohand = readfile(4, "two_hand")
lefthand = readfile(4, "left_hand")
righthand = readfile(4, "right_hand")
indexfinger = readfile(4, "index_finger")

X = np.concatenate((lefthand, righthand, indexfinger, twohand))
y = np.array([0 for x in range(156)]+ [1 for x in range(156)]+
             [2 for x in range(156)]+[3 for x in range(312)])

scaler = preprocessing.StandardScaler().fit(X)  
X_scaled = scaler.transform(X)

# random_state = seed for random generator
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X_scaled, y, test_size=0.3, random_state=0)

tuned_parameters = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4], # more parameters?
                     'C': [0.001, 0.01, 0.1 ,1, 10, 100, 1000]},
                    {'kernel': ['linear'], 'C': [0.001, 0.01, 0.1 ,1, 10, 100, 1000]}]


clf = GridSearchCV(svm.SVC(C=1), tuned_parameters, cv=10) # scoring function?
clf.fit(X_train, y_train)

print("Best parameters set found on development set:")
print
print clf.best_estimator_
print
print "Grid scores on development set:"
print
for params, mean_score, scores in clf.grid_scores_:
    print("%0.3f (+/-%0.03f) for %r"
          % (mean_score, scores.std() / 2, params))

y_true, y_pred = y_test, clf.predict(X_test)
cm = confusion_matrix(y_true, y_pred)
print
print "Confusion matrix:"
print
print(cm)
print
      
print "Confusion matrix in colour:"
print 
plt.matshow(cm)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')

for i, cas in enumerate(cm):
    for j, c in enumerate(cas):
        if c>0:
            plt.text(j-.2, i+.2, c, fontsize=20)
            
plt.show()
```
