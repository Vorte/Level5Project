```python
import numpy as np
import dataIO
from sklearn import svm, preprocessing
from sklearn.decomposition import PCA
from matplotlib import pyplot as pl
%matplotlib inline

def contains_spikes(values):
    for value in values:
        if value==65535:
            return True
    
    return False

left, right = dataIO.read_twothumb(4)

X = np.concatenate((left, right))

sizes = []
filtered=[]
count = 0
for i in range(len(X)):
    if i%52 == 0 and i<>0:
        sizes.append(count)   
        
    if not contains_spikes(X[i]):
        filtered.append(X[i])
        count += 1
    else:
        print "Spike observed at point " + str(i)

scaler = preprocessing.StandardScaler().fit(filtered)   
X_scaled = scaler.transform(filtered)

pca = PCA(n_components=2)
data = pca.fit_transform(X_scaled)

# left1 = data[:52]
# left2 = data[52:104]
# left3 = data[104:156]
# right1 = data[156:208]
# right2 = data[208:260]
# right3 = data[260:]

left1 = data[:sizes[0]]
left2 = data[sizes[0]:sizes[1]]
left3 = data[sizes[1]:sizes[2]]
right1 = data[sizes[2]:sizes[3]]
right2 = data[sizes[3]:sizes[4]]
right3 = data[sizes[4]:]
 
pl.figure(figsize=(10, 7), dpi=80)

line = pl.plot(*zip(*right1), marker='.', color='r', ls='')
line1 = pl.plot(*zip(*left1), marker='o', color='b', ls='')
line2 = pl.plot(*zip(*right2), marker='.', color='g', ls='')
line3 = pl.plot(*zip(*left2), marker='o', color='y', ls='')
line4 = pl.plot(*zip(*left3), marker='o', color='c', ls='')
line5 = pl.plot(*zip(*right3), marker='.', color='m', ls='')
      
pl.show()
```
