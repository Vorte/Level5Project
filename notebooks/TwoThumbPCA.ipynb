```python
import numpy as np
from sklearn import svm, preprocessing
from sklearn.decomposition import PCA
from matplotlib import pyplot as pl
%matplotlib inline

def readfile(userid):
    left = []
    right = []
    filenos = ["2", "6", "9"]
    createlist = lambda x: map(float, x.replace('(', '').replace(')', '').split(','))
    
    for fileno in filenos:
        filename = "/home/dimitar/Desktop/Python/experiment/results/"+str(userid)+"_"+fileno+"down.txt"
        with open(filename, "r") as f:
            lines = f.read().splitlines()
            lines = map(lambda x: x.split('\t'), lines[1:])
            map(lambda x: left.append(createlist(x[-1])) 
                if x[0]=="left" else right.append(createlist(x[-1])), lines)
        
    return np.array(left), np.array(right)

left, right = readfile(4)

X = np.concatenate((left, right))
scaler = preprocessing.StandardScaler().fit(X)   
X_scaled = scaler.transform(X)

pca = PCA(n_components = 2)
data = pca.fit_transform(X_scaled)

left1 = data[:52]
left2 = data[52:104]
left3 = data[104:156]
right1 = data[156:208]
right2 = data[208:260]
right3 = data[260:]
      
line, = pl.plot(*zip(*right1), marker='.', color='r', ls='')
line1, = pl.plot(*zip(*left1), marker='o', color='b', ls='')
line2, = pl.plot(*zip(*right2), marker='.', color='g', ls='')
line3, = pl.plot(*zip(*left2), marker='o', color='y', ls='')
line4, = pl.plot(*zip(*left3), marker='o', color='c', ls='')
line5, = pl.plot(*zip(*right3), marker='.', color='m', ls='')
      
pl.show()
```
